```python
from google.colab import files
import os
import shutil

uploaded = files.upload()
os.makedirs('/content/images', exist_ok=True)
for filename in uploaded.keys():
    shutil.move(filename, f'/content/images/{filename}')

print(f"Moved {len(uploaded)} files to /content/images")
```

```python
import cv2
from IPython.display import display, Image
```

```python
for filename in os.listdir("/content/images"):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
        try:
            img_path = f"/content/images/{filename}"
            img = cv2.imread(img_path)

            gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

            output_path = f"/content/grayscale_images/gray_{filename}"
            cv2.imwrite(output_path, gray_img)

            print(f"Converted: {filename}")
        except Exception as e:
            print(f"❌ Error processing {filename}: {str(e)}")

print("\n✅ All images processed! Check /content/grayscale_images")
```
